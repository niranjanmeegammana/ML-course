#Data Preprocessing


```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  #manage and select datasets
```

```python
#import data
url='https://raw.githubusercontent.com/Ashleshk/Machine-Learning-A-Z-hands-on-Python-And-R-in-data-Science/master/Machine%20Learning%20A-Z/Part%201%20-%20Data%20Preprocessing/Data.csv'

dataset = pd.read_csv(url)
dataset.info()
```

```python
dataset.head()
```

```python
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
le_country = preprocessing.LabelEncoder()
# Encode labels in column 'species'.
dataset['Country']= le_country.fit_transform(dataset['Country'])
le_Purchased=preprocessing.LabelEncoder()
dataset['Purchased']=le_Purchased.fit_transform(dataset['Purchased'])
```

```python
dataset
```

```python
X = dataset.iloc[:, :-1].values  #matrix of feature  : all line , all colums except last one 
y = dataset.iloc[:, 3].values   #depenedetnt variable /output
```

```python
X
```

```python
y
```

```python
# Taking care of missing data
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(X[:,1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
X
```

```python
#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()

X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

sc_y = StandardScaler()    # not required 
y_train = sc_y.fit_transform(y_train.reshape(-1,1))
```
